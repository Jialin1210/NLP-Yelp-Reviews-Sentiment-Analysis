```python

import json

# read business.json file
business_data = []
with open('/Users/Angela/Desktop/NLP_project/yelp_dataset/business.json') as f:
    for line in f:
        business_data.append(json.loads(line))

# read all the categories from business.json
result = []
for review in business_data:
    my_dict = {}
    # keep only categories
    my_dict['category'] = review.get('categories')
    result.append(my_dict)
result
```

```python
from collections import Counter

# count all categories 
total = [dic['category'] for dic in result]
output_file = open('raw_categories.txt', 'w')
for i in range(len(total)):
    output_file.writelines(str(total[i]))
    output_file.writelines('\n')
output_file.close()

all_cate = []
for sublist in total:
    cate = str.split(str(sublist), ', ')
    all_cate += cate
all_cate


count = Counter(all_cate)
cate_freq = count.most_common()
cate_freq
```

```python
print(f'total business categories: {len(cate_freq)}')
```

```python
## get all categories existed in the raw data
output_file = open('categories.txt', 'w')
for i in range(len(cate_freq)):
    output_file.writelines(cate_freq[i][0]+'\n')
output_file.close()
```

```python
## create set of lables that are restaurant-related
relevant_labels = set()
with open('restaurant_labels.txt') as labels:
    for label in labels:
        relevant_labels.add(label.rstrip())
        
relevant_labels
```

```python
len(relevant_labels)
```

```python
## get a list of business ids with restaurant-relevant categories
businessIDs_cate = []
relevantIDs = []
for d in business_data :
    categories = d.get('categories')
    cate_list = str.split(str(categories), ', ')
    new_categories = []
    for category in cate_list:
        if category in relevant_labels: # check whether category is relevant
            new_categories.append(category)
    if len(new_categories) > 0:
        my_dict = {}
        my_dict['business_id'] = d.get('business_id')
        my_dict['categories'] = new_categories
        businessIDs_cate.append(my_dict)
        relevantIDs.append(my_dict['business_id'])
relevant_ids = set(relevantIDs)
relevant_ids
```

```python
# read review.json file
review_data = []
with open('/Users/Angela/Desktop/NLP_project/yelp_dataset/review.json') as f:
    for line in f:
        review_data.append(json.loads(line))
review_data
```

```python
relevant_reviews = []
for data in review_data:
    businessid = data.get('business_id')
    if businessid in relevant_ids:
        review_dict = {}
        review_dict['business_id'] = data.get('business_id')
        review_dict['review'] = data.get('text')
        relevant_reviews.append(review_dict)
relevant_reviews
```

```python
print(f'number of reviews before filtering: {len(review_data)}')
```

```python
print(f'number of reviews after deleting all the non-restaurants reviews: {len(relevant_reviews)}')
```

```python
# check for non-retaurants for annotaion(1-200)
num = []
for entry in range(0, 200):
    if review_data[entry]['business_id'] not in relevant_ids:
        num.append(entry)
```
